# Identifikasi Outlier dalam Dataset

```python
# Menggunakan metode Boxplot

import matplotlib.pyplot as plt
import seaborn as sns

# Kolom numerik yang ingin dicek outliernya
num_cols = ['PO', 'LM', 'PK','IPM','UHH','PS','PAI','TPT','TPAK','PDRB']

# Visualisasi Boxplot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Menggunakan metode IQR

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Kuartil 1
    Q3 = data[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

Berdasarkan output tersebut data tidak memiliki outlier.

Distribusi kelas target terlihat cukup seimbang antara klasifikasi "Tidak Miskin" (0) dan "Miskin" (1), sehingga model KNN nantinya tidak terlalu bias ke salah satu kelas.

```python
# Encoding categorical variables (One-Hot Encoding)
# The 'sparse' argument has been removed in scikit-learn 1.2,
# instead, the output is always a dense array.
encoder = OneHotEncoder(sparse_output=False, drop='first') # Ensure sparse=False for compatibility
# drop='first' is used to drop the first category of each feature to avoid multicollinearity.
# handles_unknown='ignore' is used to ignore unknown categories during the transform step
encoded_categorical = encoder.fit_transform(df_cleaned[['Provinsi', 'Kab/Kota']])
# get_feature_names_out is used to get the names of the encoded features.
df_encoded = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out(['Provinsi', 'Kab/Kota']), index=df_cleaned.index)
```

```python
df_final = pd.concat([df_cleaned.drop(columns=['Provinsi', 'Kab/Kota']), df_encoded], axis=1)
```

```python
# Normalization
X = df_final.drop(columns=['KK'])
y = df_final['KK']
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
```

```python
# PCA for Dimensionality Reduction
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_scaled_df)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Finding the Best K using Cross Validation
k_values = range(1, 21)
cv_scores = [cross_val_score(KNeighborsClassifier(n_neighbors=k), X_train, y_train, cv=5, scoring='accuracy').mean() for k in k_values]
best_k = k_values[np.argmax(cv_scores)]
```

Nilai K terbaik adalah 10, berdasarkan cross-validation.

```python
# Train KNN Model
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X_train, y_train)
y_pred = knn_best.predict(X_test)
```

## Hasil Evaluasi

```python
# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
print(f"Best K: {best_k}")
print(f"Accuracy: {accuracy:.2%}")
print(f"Precision: {precision:.2%}")
print(f"Recall: {recall:.2%}")
print(f"F1 Score: {f1:.2%}")
print(f"ROC AUC: {roc_auc:.2%}")
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", classification_rep)
```

Hasil evaluasi model KNN dengan K=10 pada data uji:

Akurasi: 94%  Model cukup baik dalam memprediksi kelas dengan benar.
Precision: 75%  Dari semua yang diprediksi sebagai "Miskin", 75% benar.
